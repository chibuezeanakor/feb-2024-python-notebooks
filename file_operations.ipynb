### To read a file

```python
file = open("sample.txt", 'r') # Read mode operation
content = file.read()
print(content)
file.close() # Important operation
```

### To write to a file

```python
file = open("output.txt", 'w')
file.write("Hello again, World!")
file.close()
```

### Append to a file

```python
file = open("log.txt", 'a')
file.write("Another access at 10:38 AM\n")
file.close()
```

```python
file = open("log.log", 'w')
file.write("Created at 11:03 AM")
file.close()
```

### Reading lines from a file

```python
file = open("data.txt", 'r')
lines = file.readlines() # Reads all lines into a list
for line in lines:
    print(line.strip())
    
file.close()
```

```python
text = "   Hello World!"
cleanText = text.strip()
print(cleanText)
```

### Open and close the file automatically by using "with" keyword

```python
with open("sample.txt", 'r') as file:
    content = file.read()
    print(content)
```

### Open a binary image file

```python
with open("penguin.png", 'rb') as file:
    content = file.read()
#    print(f"The files first byte: {content[0]}")
    print(content)
```
